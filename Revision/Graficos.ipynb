```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import calendar
import matplotlib.dates as mdates
```

```python
datos = pd.read_excel('datos.xlsx')
```

```python
meses = {'Enero':1, 'Febrero':2, 'Marzo':3, 'Abril':4, 'Mayo':5, 'Junio':6, 'Julio':7, 'Agosto':8, 'Septiembre':9, 'Octubre':10, 'Noviembre':11, 'Diciembre':12}
```

```python
datos['Mes'] = datos['Mes'].map(meses)
```

```python
datos = datos.rename(columns={'Mes': 'month', 'AÃ±o':'year'})
```

```python
datos['fecha'] = pd.to_datetime(datos[['year', 'month']].assign(Day=1))
```

```python
sns.lineplot(x=datos['fecha'], y=datos['Reserva_Neta'], label='Reserva Neta Estimada')
sns.lineplot(x=datos['fecha'], y=datos['Reserva_Neta_Real'], label='Reserva Neta Observada')
plt.xlabel('Fecha')
plt.ylabel('Reserva en millones de colones')
plt.xticks(rotation=45)
plt.savefig('ReservaNeta_Real.png', format='pdf', bbox_inches='tight')
```

```python
sns.lineplot(x=datos['fecha'], y=datos['TC_DOLAR'], label='Tipo de cambio colones por dolar')
plt.xlabel('Fecha')
plt.ylabel('Colones por un dolar')
plt.xticks(rotation=45)
plt.savefig('TC.png', format='pdf', bbox_inches='tight')
```
